# Column SibSp dan Column Parch
we will do data manipulation

manipulation here is not to change the data value therapy to make this data easier to read by machine

Column SibSp(sibling Spouse) means a column that states the number of siblings or the number of partners brought by the Passanger column Parch (Parent Child) means column which states the number of parents or the number of children brought by the Passanger

we will create a new column that displays whether he is alone or with his family

```python

df['Alone']=df['SibSp']+df['Parch']
```

```python


df['Alone'][df['Alone']>0]='With Family'
df['Alone'][df['Alone']==0]='Without Family'
```
